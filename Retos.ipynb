## Implementa el algoritmo de la Criba de Erastótenes para encontrar número primos. Encuentra los números primos entre los 10.000 primeros números naturales

```python
# Criba de Erastótenes
# Escribe a ti tu código
def criba_eratostenes(n):
    # Creamos una lista para marcar los números compuestos
    # Inicialmente asumimos que todos los números son primos
    es_primo = [True] * (n+1)
    es_primo[0] = es_primo[1] = False

    # Aplicamos la Criba de Eratóstenes
    for i in range(2, int(n**0.5) + 1):
        if es_primo[i]:
            # Si i es primo, marcamos todos los múltiplos de i como compuestos
            for j in range(i*i, n+1, i):
                es_primo[j] = False

    # Retornamos la lista de números primos
    primos = [num for num, is_prime in enumerate(es_primo) if is_prime]
    return primos

# Encontrar números primos hasta los primeros 10,000 números naturales
primos_hasta_10000 = criba_eratostenes(10000)
print(primos_hasta_10000)
```

## Opción 2: Implementa una solución del problema de las Torres de Hanoi para un número arbitrario n de fichas

```python

# Torres de Hanoy
# Escribe aquí tu código

def hanoi(n, source, target, auxiliary):
    if n == 1:
        print(f"Mover disco 1 de {source} a {target}")
        return

    hanoi(n-1, source, auxiliary, target)
    print(f"Mover disco {n} de {source} a {target}")
    hanoi(n-1, auxiliary, target, source)

num_disks = 4
hanoi(num_disks, 'A', 'C', 'B')


```

```python
import numpy as np
import math


# Definimos la funcion
f = lambda X: math.sin(1/2 * X[0]**2 - 1/4 * X[1]**2 + 3) * math.cos(2*X[0] + 1 - math.exp(X[1]))

def gradient_descent(initial_point, learning_rate, iterations):
    x, y = initial_point
    for _ in range(iterations):
        gradient_x = x * math.cos(0.5 * x**2 - 0.25 * y**2 + 3) * math.cos(2 * x + 1 - math.exp(y)) - 2 * math.sin(0.5 * x**2 - 0.25 * y**2 + 3) * math.sin(2 * x + 1 - math.exp(y))
        gradient_y = 0.5 * y * math.cos(0.5 * x**2 - 0.25 * y**2 + 3) * math.cos(2 * x + 1 - math.exp(y)) + math.exp(y) * math.sin(0.5 * x**2 - 0.25 * y**2 + 3) * math.sin(2 * x + 1 - math.exp(y))

        x -= learning_rate * gradient_x
        y -= learning_rate * gradient_y

    return x, y

# Parámetros
initial_point = [1.0, 1.0]
learning_rate = 0.001
iterations = 5000

# Descenso del gradiente
result_x, result_y = gradient_descent(initial_point, learning_rate, iterations)

# Resultado
result = f([result_x, result_y])

print("Resultado:")
print(f"f({result_x}, {result_y}) = {result}")
```

Implemente el código del descenso por gradiente para la regresión lineal.

```python
import numpy as np
import matplotlib.pyplot as plt

def generar_datos(num_muestras, m_real, b_real, desviacion_estandar):
    X = np.random.rand(num_muestras)
    ruido = np.random.randn(num_muestras) * desviacion_estandar
    y = m_real * X + b_real + ruido
    return X, y

def costo(y_true, y_pred):
    N = len(y_true)
    return np.sum((y_true - y_pred) ** 2) / (2 * N)

def descenso_gradiente(X, y, tasa_aprendizaje, num_iteraciones):
    N = len(y)
    m = 0.0
    b = 0.0
    historial_costo = []

    for _ in range(num_iteraciones):
        y_pred = m * X + b

        d_m = -2 * np.sum((y - y_pred) * X) / N
        d_b = -2 * np.sum(y - y_pred) / N

        m = m - tasa_aprendizaje * d_m
        b = b - tasa_aprendizaje * d_b

        # Costo en cada iteración
        costo_actual = costo(y, y_pred)
        historial_costo.append(costo_actual)

    return m, b, historial_costo

# Generacio datos
m_real = 2.5
b_real = 5.0
desviacion_estandar = 1.0
num_muestras = 100
X, y = generar_datos(num_muestras, m_real, b_real, desviacion_estandar)

# Parámetros
tasa_aprendizaje = 0.1
num_iteraciones = 100

# Descenso por gradiente
m, b, historial_costo = descenso_gradiente(X, y, tasa_aprendizaje, num_iteraciones)

print(f"Pendiente (m): {m}")
print(f"Intersección (b): {b}")

# Visualizamos la evolución del costo en cada iteración
plt.plot(range(num_iteraciones), historial_costo)
plt.xlabel('Iteraciones')
plt.ylabel('Costo')
plt.title('Evolución del costo durante el descenso por gradiente')
plt.show()

# Visualizamos los datos y la recta ajustada por el modelo
plt.scatter(X, y, label='Datos')
plt.plot(X, m * X + b, color='red', label='Recta ajustada')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.show()
```
